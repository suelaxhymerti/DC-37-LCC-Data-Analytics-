Research Question: Race was chosen as the comparison group to analyze differences in diabetes hospital data.

```python
import pandas as pd
import seaborn as sns
url = "https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv"
df = pd.read_csv(url)
df.head()
```

```python
df['race'] = df['race'].str.strip()
df['race'] = df['race'].replace('?', pd.NA)
```

```python
df.dropna(subset=['race'], inplace=True)
```

```python
df["race"].value_counts(dropna=False)
```

```python

df['race'].value_counts()
```

```python
df.groupby('race')['time_in_hospital'].mean()
```

```python
sns.boxplot(x='race', y='time_in_hospital', data=df)
```

```python
sns.countplot(data=df, x='race')
```

The analysis reports differences in average hospital stay across racial groups. Caucasian and African American patients had slightly different average lengths of stay.
